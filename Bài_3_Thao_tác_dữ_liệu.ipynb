# Loc
- Cú pháp là: loc[dòng, cột] phân cách bởi dấu phẩy
  + dòng: có thể là [chỉ số dòng] hoặc [khoảng dòng (chỉ số đầu: chỉ số cuối)] hoặc là [danh sách dòng]
  + cột: có thể là [nhãn của cột] hoặc [danh sách cột] hoặc là [khoảng cột (nhãn đầu: nhãn cuối)]
  + Trong trường hợp không có dấu phẩy và chỉ có 1 thành phần thì được hiểu là đang truy suất dòng.
- Mục tiêu: Lọc ra một vùng của data frame, sử dụng nhãn
- Hỗ trợ cả các phép so sánh

# Hàm filter
- Cú pháp filter(items=None, like=None, regex=None, axis=None)
- Ứng dụng: để lấy ra một tập con của dataframe, kết quả trả ra thì có thể là DataFrame
- axis = 0 là lọc dữ liệu theo dòng
- axis = 1 lọc dữ liệu theo cột
- regex: cú pháp hỗ trợ tìm kiếm linh động theo tên (chỉ số) dòng hoặc cột
- like: tìm kiếm trên nhãn nhãn của chỉ số dòng hoặc cột

- Trong trường hợp không dùng regex, hơi giống toán tử ngoặc vuông và các hàm ở trên


```python
df.filter(items=['Country', 'Continent'], axis=1)
```

```python
df.filter(items=['Country', 'Continent'], axis=1)[4:6]
```

# Hàm query
Chỉ lọc được dữ liệu theo cột

```python
df[df['Country'].str.contains('ria$')]
```

```python
import numpy as np
```

```python
arr = np.random.randint(int(df['GDP (millions of US$)'].min())-100, int(df['GDP (millions of US$)'].max())+10000, size=(df.shape[0]))
```

```python
arr
```

```python
df['randint']=pd.Series(arr)
```

```python
df.head()
```

```python
df = df.rename(columns={"GDP (millions of US$)":"GDP"})
```

```python
df.info()
```

```python
df.head()
```

```python
df_4 = df.query('GDP > randint')
```

```python
df_4.info()
```

```python
df_4
```

# Có rất nhiều lệnh tương đương query

```python
df_5 = df[df.GDP>df.randint]
df_5
```
