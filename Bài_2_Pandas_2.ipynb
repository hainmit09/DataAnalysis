# - hàm drop
DataFrame.drop(self, labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

- Xóa dòng
  + Cách 1 axis= 0, labels=các nhãn cần xóa
  + Cách 2 index=labels

** Labels có thể là 1 nhãn hoặc danh sách nhãn của các dòng


```python
df.drop(index=7379, inplace=True)
```

```python
df.tail()
```

```python
df.drop(axis=0, labels=[7376, 7375], inplace=True)
```

```python
df.tail()
```

- Xóa cột: 
  + Cách 1: labels= cột hoặc danh sách cột kết hợp với axis =1, 
  + Cách 2 columns=['tên cột 1',..]
